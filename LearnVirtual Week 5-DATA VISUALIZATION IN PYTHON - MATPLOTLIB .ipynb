## 1. MAKING BASIC PLOTS USING THE MATLAB STYLE IN MATPLOTLIB

```python
#making a simple plot with matplotlib using a simple data
#A simple line graph
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
plt.plot(x,y)
plt.show()
```

### Adding Labels to the Plot

```python
#addiing more arguments to the plot
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
plt.plot(x,y)
plt.title('A plot of X vs Y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

### Changing font size

```python
#addiing more arguments to the plot
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
plt.plot(x,y)
plt.title('A plot of X vs Y', fontsize=18)
plt.xlabel('X-axis', fontsize=16)
plt.ylabel('Y-axis', fontsize=16)
plt.show()
```

### Adjusting tick size

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
plt.plot(x,y)
plt.title('A plot of X vs Y', fontsize=18)
plt.xlabel('X-axis', fontsize=16)
plt.ylabel('Y-axis', fontsize=16)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

### Adding markers

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
plt.plot(x,y, marker = '.', markersize = 10)
plt.title('A plot of X vs Y', fontsize=18)
plt.xlabel('X-axis', fontsize=16)
plt.ylabel('Y-axis', fontsize=16)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

## Let's make more Plots to see other functionalities

```python
#Making a multiple plots and adding a legend
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
z = [1, 3, 5, 7, 9]
plt.plot(x,y, marker='v')
plt.plot(x,z, marker='*')
plt.title('A plot of X Vs Y and X Vs Z')
plt.xlabel('X-axis')
plt.ylabel('Y-axis and Z-axis')
plt.legend(['X vs z', 'X vs y'])
plt.show()
```

## Moving the Position of the Legend

```python
plt.plot(x,y, marker='v')
plt.plot(x,z, marker='*')
plt.title('A plot of X Vs Y and X Vs Z')
plt.xlabel('X-axis')
plt.ylabel('Y-axis and Z-axis')
plt.legend(['X vs z', 'X vs y'],loc=(1.02,0), borderaxespad=0, fontsize = 20)
plt.show()
```

```python
plt.plot(x,y, marker='v')
plt.plot(x,z, marker='*')
plt.title('A plot of X Vs Y and X Vs Z')
plt.xlabel('X-axis')
plt.ylabel('Y-axis and Z-axis')
plt.legend(['X vs z', 'X vs y'],loc=('best'), borderaxespad=0, fontsize = 20)
plt.show()
```

## Changing Color
The `c` parameter accepts strings.

string | color
--- | ---
'b' | blue
'blue' | blue
'g' | green
'green' | green
'r' | red
'red' | red
'c' | cyan
'cyan' | cyan
'm' | magenta
'magenta' | magenta
'y' | yellow
'yellow' | yellow
'k' | black
'black' | black
'w' | white
'white' | white

The parameter also accepts hex strings. For instance, green is '#008000'. Additionally you can use rgb tuples. 

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
z = [1, 3, 5, 7, 9]
plt.plot(x,y, c='m')
plt.plot(x,z, c='c')
plt.title('A plot of X Vs Y and X Vs Z')
plt.xlabel('X-axis')
plt.ylabel('Y-axis and Z-axis')
plt.legend(['X vs z', 'X vs y'])
plt.show()
```

# Bar Charts
- The bar() function in Matplotlib takes two arguments **(x,y)**
- While X = a category
- Y = A numeric value

```python
data = {'Tutors': ['Aminah', 'Adeola', 'Fatimah', 'Stephen', 'Femi', 'Isreal', 'Olayinka'],
        'Participants': ['joseph', 'Benjamin', 'Emmanuel', 'Kafee', 'Onyinye', 'Dominique', 'Andiwa'],
        'tutors_gender': ['Female', 'Female', 'Female', 'Male', 'Male', 'Male', 'Male'],
        'participant_gender': ['Male', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female'], 
        'Age':[25,29,24,34,32,27,30]}
Data = pd.DataFrame(data)
Data.head()
```

```python
plt.style.use('classic')
x = Data.Tutors
y = Data.Age

plt.bar(x,y, color='chocolate')
plt.xlabel('Tutors-Gender')
plt.ylabel('Age-range')
```

```python
#Creating an Horizontal barchart
plt.style.use('classic')
x = Data.Tutors
y = Data.Age

plt.barh(x,y, color='tomato')
plt.xlabel('Tutors-Gender')
plt.ylabel('Age-range')
```

```python
#Adding arguments
plt.style.use('seaborn-muted')
x = Data.Tutors
y = Data.Age

plt.barh(x,y, color='indianred',align='center')
plt.xlabel('Tutors-Gender')
plt.ylabel('Age-range')
```

## Saving Your Plots to Your Local Computer


```python
plt.style.use('seaborn-pastel')
x = Data.Tutors
y = Data.Age

plt.barh(x,y, color='indianred',align='center')
plt.xlabel('Tutors-Gender')
plt.ylabel('Age-range')
plt.grid(True)
plt.savefig('A simple Hist', dpi=300)
```

# Object-Oriented Style Plotting 

#### *The difference is in the syntax*

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y)
```

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]
z = [1, 3, 5, 7, 9]
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y, c='m')
ax.plot(x,z, c='c')
ax.set(title='A plot of X Vs Y and X Vs Z', xlabel='X-axis',ylabel='Y-axis')
ax.set_legend('')
ax.set_title()
ax.set_xlabel()
#plt.legend(['X vs z', 'X vs y'], loc=('best'))
```
